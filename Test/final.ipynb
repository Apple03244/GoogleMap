```python
from notion_client import Client
from typing import List,Dict

class Notion(Client):
    def __init__(self, options = None, client = None, **kwargs):
        super().__init__(options, client, **kwargs)

    def __decode_properites__(self,input:dict):
        if input["type"] in ["rich_text","text","title"]:
            if input[input["type"]]:
                return input[input["type"]][0]["text"]["content"]
            else:
                return None
        elif input["type"] in ["relation"]:
            return None
        else:
            return input[input["type"]]

    def search_properites(self,database_id:str):
        main_result=self.search(filter={"property": "object", "value": "page"})["results"]
        tabling_result={}
        if not main_result:
            print("api_key 와 database_id 를 확인하세요")

        else:
            properties_frame={}
            for result in main_result:
                if result["parent"]["database_id"].replace("-","").strip()==database_id:
                    sub_frame=result["properties"]

                    for key in sub_frame:
                        if key not in properties_frame:
                            properties_frame[key]=[self.__decode_properites__(sub_frame[key])]
                        else:
                            properties_frame[key].append(self.__decode_properites__(sub_frame[key]))
            return properties_frame
                
    def upload_to_notion(self,database_id,infos:List[Dict]):
        for sub_info in infos:
            page=self.pages.create(
                parent={"database_id": database_id},
                properties=sub_info
            )
```

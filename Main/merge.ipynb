```python
from notion_client import Client
from typing import List,Dict
from typing import Union

class Notion(Client):
    def __init__(self, options = None, client = None, database_ids:list=[],**kwargs):
        super().__init__(options, client, **kwargs)

    def __property_make(property_type:str,value:Union[int,float,str]):
        if property_type in ["text","rich_text","title"]:
            return {property_type : [{"text": {"content": value}}]}
        elif property_type in ["select",'created_time','relation']: # 차후 업데이트
            pass
        else:
            return {property_type : value}
    # 편의성
    def __frame(self,database_id):
        frame_base=self.search(filter={"property": "object", "value": "page"})["results"]
        try:
            frame_base=[fb["properties"] for fb in frame_base 
                             if fb["parent"]["database_id"].replace("-","")==database_id][0]
        except ValueError as ve:
            print(f"{ve} : database_id 를 확인하세요")

        __properties__={"parent":{"database_id": database_id}}
        type_info={}
        for base in frame_base:
            type_info.update({base:frame_base[base]["type"]})
        __properties__.update({"type_info":type_info})

        self.__properties__=__properties__
        return __properties__

    def upload_to_notion(self,database_id,infos:List[Dict]):
        '''
        노션 db에 업로드(차후 중복 채크 필요)
        '''
        for sub_info in infos:
            page=self.pages.create(
                parent={"database_id": database_id},
                properties=sub_info
            )
    def load_db(self,database_id,**kwargs):
        self.__frame(database_id)
        return self.__properties__
```
